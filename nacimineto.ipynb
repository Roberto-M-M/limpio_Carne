Eliminaremos las columnas de 2023 y Unnamed: 68 por ser datos muy nuevos y incompletos.

```python
nacimiento1 = nacimiento.drop(columns=['2023', 'Unnamed: 68'])
nacimiento1
```

Ahora eliminaremos las columnas de Country Code, Indicator Name y Indicator Code por que estos datos no los necesitamos

```python
nacimiento2 = nacimiento1.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'])
nacimiento2
```

```python
#Primero podemos inspeccionar si hay valores únicos con el siguiente comando:

nacimiento2.nunique()
```

Buscamos a chile en este data set

```python
chile_row = nacimiento2[nacimiento2['Country Name'].str.contains('Chile')]
chile_row
```

Ahora cambiamos el tipo de dato para trabajar mas facil a la hora de hacer los graficos.

```python
#Primero debemos cambiar el tipo de dato a un tipo categórico para que la función logre entender la transformación
nacimiento2['Country Name'] = nacimiento2['Country Name'].astype('category')
nacimiento2.dtypes
```

```python
df = nacimiento2
print(df.shape)
df.head()
```

```python
# Leer el archivo CSV
df = nacimiento2

# Filtrar las filas donde la columna 'Entity' es 'Chile'
df_chile = df[df['Country Name'] == 'Chile']

# Guardar el resultado en un nuevo archivo CSV
df_chile.to_csv('nacimiento.csv', index=False)

print("Se ha creado el archivo 'chile_datos.csv' con los datos filtrados.")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model

# Filtrar el DataFrame para obtener solo los datos de Chile
chile_df = df[df['Country Name'] == 'Chile']

# Extraer los años desde la primera fila del DataFrame
years = chile_df.columns[5:]

# Seleccionar solo algunos años para mostrar en el gráfico
step = 5
selected_years = years[::step]

# Crear la gráfica lineal
plt.plot(years, chile_df.iloc[:, 5:].values.flatten(), marker='o')  # Seleccionar valores de población
plt.xlabel('Año')
plt.ylabel('Población')
plt.title('Población en Chile por año')
plt.xticks(selected_years, rotation=45)  # Rotar las etiquetas del eje x para mayor claridad
plt.grid(True)  # Añadir una cuadrícula para una mejor visualización
plt.show()
```
